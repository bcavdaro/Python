```python
note = int(input("Sinavdan aldiginiz notu girin"))

if 0 <= note <= 100:
    if note >= 80:
        harf_notu = "A"
    elif note >= 65:
        harf_notu = "B"
    elif note >= 55:
        harf_notu = "C"
    elif note >= 50:
        harf_notu = "D"
    else:
        harf_notu = "F"
    print(f"Aldiginiz puana denk gelen harf notunuz : {harf_notu}")
else:
    print("hatali not girisi yaptiniz")
```
